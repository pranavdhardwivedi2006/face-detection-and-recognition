```python
import cv2
```

```python
haar_features = cv2.CascadeClassifier('../models/haarcascade_frontalface_default.xml')
```

```python
img = cv2.imread('images/01_msd.jpeg')
img.shape
```

```python
haar_features.detectMultiScale(img)
```

```python
faces = haar_features.detectMultiScale(img, scaleFactor = 1.3)
for (x, y, w, h) in faces :
    # img - on which we want to draw rectangle
    # params : image, top-left point, (width, height), colour of face box, border width of box
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 0, 255), 1)
cv2.imshow('output', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)
while True :
    flag, frame = cap.read()
    if not flag :
        print('Camera not accessible...')
        break
    faces = haar_features.detectMultiScale(frame, scaleFactor = 1.3)
    for (x, y, w, h) in faces :
        # img - on which we want to draw rectangle
        # params : image, top-left point, (width, height), colour of face box, border width of box
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 0, 255), 2)
    cv2.imshow('output', frame)
    key = cv2.waitKey(5) & 0xFF
    if key == 27 :
        break
cap.release()
cv2.destroyAllWindows()
```
